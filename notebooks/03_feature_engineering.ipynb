```python
import pandas as pd
import numpy as np

df = pd.read_csv('cleaned_zameen.csv')

df['log_price'] = np.log1p(df['price'])

df['area_bed_ratio'] = df['area_size'] / df['bedrooms'].replace(0,1)

df['luxury_flag'] = (df['price'] > df['price'].quantile(0.90)).astype(int)

df.to_csv('featured_zameen.csv', index=False)
```
