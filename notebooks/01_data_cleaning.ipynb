```python
import pandas as pd
import numpy as np

df = pd.read_csv('zameen.csv')

print("Initial Shape:", df.shape)

df = df.drop_duplicates()

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

drop_cols = ['page_url', 'agent', 'property_id', 'location_id']
df = df.drop(columns=drop_cols, errors='ignore')

df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['area_size'] = pd.to_numeric(df['area_size'], errors='coerce')
df['baths'] = pd.to_numeric(df['baths'], errors='coerce')
df['bedrooms'] = pd.to_numeric(df['bedrooms'], errors='coerce')

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

num_cols = ['price','area_size','baths','bedrooms','latitude','longitude']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

cat_cols = ['property_type','location','city','province_name','purpose','area_type','area_category']
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

df = df[(df['price'] > 100000) & (df['price'] < df['price'].quantile(0.99))]
df = df[(df['area_size'] > 0) & (df['area_size'] < df['area_size'].quantile(0.99))]

df.reset_index(drop=True, inplace=True)

print("Final Shape:", df.shape)

df.to_csv('cleaned_zameen.csv', index=False)
```
