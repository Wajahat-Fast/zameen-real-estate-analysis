```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('cleaned_zameen.csv')

plt.figure()
df['price'].hist(bins=60)
plt.title('Property Price Distribution')
plt.show()

plt.figure()
sns.scatterplot(x='area_size', y='price', data=df.sample(5000))
plt.title('Area vs Price')
plt.show()

plt.figure()
sns.barplot(x='bedrooms', y='price', data=df)
plt.title('Bedrooms vs Price')
plt.show()

plt.figure()
sns.barplot(x='baths', y='price', data=df)
plt.title('Bathrooms vs Price')
plt.show()

top_city = df.groupby('city')['price'].mean().sort_values(ascending=False).head(10)

plt.figure()
top_city.plot(kind='bar')
plt.title('Top 10 Expensive Cities')
plt.show()

plt.figure()
sns.scatterplot(x='longitude', y='latitude', hue='price', data=df.sample(5000))
plt.title('Geo Price Distribution')
plt.show()
```
